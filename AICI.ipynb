```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

PORTFOLIO_CSV = "AICI_final_portfolio_with_weights.csv"

pf = pd.read_csv(PORTFOLIO_CSV)

cols_lower = {c.lower(): c for c in pf.columns}

ticker_col = cols_lower.get("ticker", None)
if ticker_col is None:
    # fall back: first column
    ticker_col = pf.columns[0]

weight_col = cols_lower.get("final_weight", None)
if weight_col is None:
    # fall back: second column
    weight_col = pf.columns[1]

pf_tickers = pf[ticker_col].astype(str).str.strip().tolist()
pf_weights = pf[weight_col].astype(float).values

print("Original tickers:", pf_tickers)
print("Original weight sum:", pf_weights.sum())


start_date = "2018-01-01"
end_date   = "2025-12-03"  # or today

# Benchmarks
bench_tickers = ["SOXX", "SKYY", "WTAI"]


price_df = pd.DataFrame()
used_tickers = []
used_weights = []

for t, w in zip(pf_tickers, pf_weights):
    print(f"Downloading: {t}")
    s = yf.download(t, start=start_date, end=end_date)

    if s.empty:
        print(f"  -> No data for {t}, skipping.")
        continue

    # Take Adj Close if available, else Close
    if "Adj Close" in s.columns:
        col = s[["Adj Close"]].copy()
    elif "Close" in s.columns:
        col = s[["Close"]].copy()
    else:
        print(f"  -> No usable price column for {t}, skipping.")
        continue

    col.columns = [t]         # rename column to ticker
    price_df = pd.concat([price_df, col], axis=1)

    used_tickers.append(t)
    used_weights.append(w)

print("\nUsed tickers:", used_tickers)
print("Weight sum BEFORE renormalizing:", sum(used_weights))

if len(used_tickers) == 0:
    raise RuntimeError("No valid price data for any portfolio ticker. Stop here.")

# Renormalize weights to 1 after skipping any failed tickers
used_weights = np.array(used_weights, dtype=float)
used_weights = used_weights / used_weights.sum()
print("Weight sum AFTER renormalizing:", used_weights.sum())

print("Price DF shape:", price_df.shape)


q_prices = price_df.resample("Q").last()
q_rets   = q_prices.pct_change().dropna(how="all")

w_series = pd.Series(used_weights, index=used_tickers)
port_qrets = (q_rets[used_tickers] * w_series).sum(axis=1)
port_qrets.name = "AICI"


bench_price_df = pd.DataFrame()

for bt in bench_tickers:
    print("Downloading benchmark:", bt)
    s = yf.download(bt, start=start_date, end=end_date)

    if s.empty:
        print(f"  -> No data for benchmark {bt}, skipping.")
        continue

    if "Adj Close" in s.columns:
        col = s[["Adj Close"]].copy()
    elif "Close" in s.columns:
        col = s[["Close"]].copy()
    else:
        print(f"  -> No usable price column for benchmark {bt}, skipping.")
        continue

    col.columns = [bt]
    bench_price_df = pd.concat([bench_price_df, col], axis=1)

# Align benchmark sample to portfolio period
bench_q = bench_price_df.resample("Q").last().pct_change().dropna(how="all")
bench_q = bench_q.loc[port_qrets.index.min():port_qrets.index.max()]


def perf_stats(q_returns, name="AICI", rf=0.0):
    """
    q_returns: quarterly return series
    rf: annual risk-free (we leave at 0 for simplicity)
    """
    q_returns = q_returns.dropna()
    if len(q_returns) == 0:
        return pd.Series(
            {"CAGR": np.nan, "Volatility": np.nan,
             "Sharpe": np.nan, "MaxDrawdown": np.nan},
            name=name
        )

    cum = (1 + q_returns).cumprod()
    n_quarters = len(q_returns)
    n_years = n_quarters / 4.0

    final_val = cum.iloc[-1]
    cagr = final_val ** (1 / n_years) - 1

    vol = q_returns.std() * np.sqrt(4)
    sharpe = (cagr - rf) / vol if vol > 0 else np.nan

    running_max = cum.cummax()
    dd = cum / running_max - 1
    max_dd = dd.min()

    return pd.Series(
        {"CAGR": cagr, "Volatility": vol,
         "Sharpe": sharpe, "MaxDrawdown": max_dd},
        name=name
    )


stats_list = [perf_stats(port_qrets, "AICI")]

for bt in bench_tickers:
    if bt in bench_q.columns:
        stats_list.append(perf_stats(bench_q[bt], bt))
    else:
        print(f"WARNING: No quarterly data for benchmark {bt}")

stats = pd.DataFrame(stats_list)

print("\n==============================")
print("Performance (Annualized)")
print("==============================")
print(stats.applymap(lambda x: round(x, 4)))


cum_df = pd.DataFrame(index=port_qrets.index)
cum_df["AICI"] = (1 + port_qrets).cumprod()

for bt in bench_tickers:
    if bt in bench_q.columns:
        cum_df[bt] = (1 + bench_q[bt]).cumprod()


plt.figure(figsize=(10, 6))
for col in cum_df.columns:
    plt.plot(cum_df.index, cum_df[col], label=col)

plt.title("Cumulative Growth of $1  AICI vs Benchmarks (Quarterly)")
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


stats.to_csv("AICI_backtest_stats.csv", index=True)
cum_df.to_csv("AICI_backtest_cumulative.csv", index=True)

print("\nSaved:")
print(" - AICI_backtest_stats.csv")
print(" - AICI_backtest_cumulative.csv")
print("\nDone.")
```
